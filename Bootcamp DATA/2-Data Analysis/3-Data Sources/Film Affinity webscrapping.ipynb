```python
import pandas as pd
from selenium import webdriver

driver = r'C:\Users\angel\Desktop\Google Chrome Driver\chromedriver'

options = webdriver.ChromeOptions()
#options.add_argument('headless') #abre chrome para descargar cosas pero no lo ves

driver = webdriver.Chrome(executable_path = driver, options = options)

url = 'https://www.filmaffinity.com'
driver.get(url)

elementstag=driver.find_elements_by_tag_name('button')
boton_aceptar = elementstag[2]
boton_aceptar.click()

estrenos = driver.find_element_by_xpath('/html/body/div[5]/table/tbody/tr/td[1]/div/div[2]/ul/li[3]/a')
estrenos.click()

wrapper = driver.find_element_by_id('main-wrapper-rdcat')
peliculas = wrapper.find_elements_by_class_name("top-movie")
url_films = list()

for pelicula in peliculas:
    url_films.append(pelicula.find_element_by_tag_name('a').get_attribute('href'))
    
url_films
asdlist =[]

for pelis in url_films:
    driver.get(pelis)
    lista_2 = driver.find_element_by_class_name('movie-info').find_elements_by_tag_name('dd')
    lista_1 = driver.find_element_by_class_name('movie-info').find_elements_by_tag_name('dt')
    lista_vacia =[]
    lista_vacia2 = []
    
    for i in lista_1:
        lista_vacia.append(i.text)
        
    for i in lista_2:
        lista_vacia2.append(i.text)
    asdlist.append(pd.DataFrame([lista_vacia2], columns = lista_vacia))
    
dfs = [df.reset_index() for df in asdlist]
df2 = (pd.concat(dfs, axis=0))

newdf = df2.drop(['index'], axis=1)
newdf.index = range(1,9)
newdf
```
