```python
query = """
SELECT country, COUNT(*)
FROM customers
WHERE country == 'Spain' or country == 'Brazil'
GROUP BY country;
"""
cursor.execute(query)
cursor.fetchall()
```

```python
# 1
query = """
SELECT c.Firstname, i.invoiceid, i.invoicedate, i.billingcountry
FROM customers c
JOIN invoices i ON c.customerId == i.customerId
WHERE c.country == 'Brazil';
"""
cursor.execute(query)
cursor.fetchall()
```

```python
#2
query = """
SELECT i.*
FROM customers c
JOIN invoices i ON c.customerId == i.customerId
WHERE c.country == 'Brazil';
"""
cursor.execute(query)
cursor.fetchall()
```

```python
#3
query = """
SELECT e.firstname || ' ' || e.lastname as 'Agente de ventas',i.*
FROM customers c
JOIN invoices i ON c.customerId == i.customerId
JOIN employees e ON c.supportrepid == e.employeeid;
"""
cursor.execute(query)
cursor.fetchall()
```

```python
#4
query = """
SELECT c.firstname || ' ' || c.lastname as 'Nombre cliente', c.country, e.firstname || ' ' || e.lastname as 'Agente de ventas', i.total
FROM customers c
JOIN invoices i ON c.customerId == i.customerId
JOIN employees e ON c.supportrepid == e.employeeid;
"""
cursor.execute(query)
cursor.fetchall()
```

```python
#5
query = """
SELECT  i.invoiceid, it.quantity, t.name
FROM invoices i 
JOIN invoice_items it ON i.invoiceid == it.invoiceid
JOIN tracks t ON it.trackid == t.trackid
;
"""
cursor.execute(query)
cursor.fetchall()
```

```python
#6
query = """
SELECT  t.name as 'Nombre', mt.name as 'Formato', a.title as 'Album', g.name as 'Genero'
FROM tracks t 
JOIN albums a ON t.albumid == a.albumid
JOIN genres g ON t.genreid == g.genreid
JOIN media_types mt ON t.mediatypeid == mt.mediatypeid
;
"""
cursor.execute(query)
cursor.fetchall()
```
